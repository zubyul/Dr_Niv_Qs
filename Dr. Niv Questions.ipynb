## 1b written
If someone has a learning rate of n_pos > or < n_neg, their percieved value of a consistent reward will not change, as the predicted error is zero. Similarly, the percieved value of the zero condition will also remain zero, and not go lower.

However, if someone has n_pos > n_neg and they encounter the condition with the constant prediction error, they will show a constant growth in value for the stimulus, beyond the value of the condition with a constant reward. If they have n_neg >n_pos, they will show a constant decline in value for the stimulus, beyond the value of the condition with a constant 0 value.

I believe this implies there is a missing variable (or I did something wrong). Despite the existance of gambling addiction, it doesn't seem correct that the percieved value of an output of (0, 10, 10, 0) would be valued higher than (10, 10, 10, 10). In the context of risk aversion, (0, 10, 10, 0) being valued less that (0, 0, 0, 0). I presume this is because the code does not account for the change in prediction error decreasing in value as it becomes 'predictably unpredictable', as the value is always either 0 or 10. Such a modification would still allow for rarer higher rewards or higher punishments to still effect the output, creating stronger addiction or aversion.

option 1: 0.01$
option 2: 0$

they would mentally set the value of option 1 > option 2 and proceed to only choose option 1.

The 1 in 100 people who first get the output:

option 1: 0.01$
option 2: 1$

Will mentally set the value of option 2 >> option 1. In this case, people with n_neg>n_pos will devalue option 2 faster than people with n_pos>n_neg. People with n_pos>n_neg, depending on their relative postive/negative learning rates and rate of success, are likely to value option 2 > option 1 indefinitely.
